## Relabel Data

```python
# obtain all of the file paths from the CPhoenix UOF Data folder and sort them into a data frame by subject

path_folder = 'CPhoenix UOF Data/'
data_filepaths = glob.glob(f'{path_folder}*__.csv', recursive = True)
data_dict_filepaths = glob.glob(f'{path_folder}*-data-dictionary.csv', recursive = True) #

file_subjects = []

for path_data, path_dict in zip(data_filepaths, data_dict_filepaths):
    
    file_subject = path_data.rsplit('\\', 1)[-1][5:-6] #the range would be [5:-20] for the data dicts
    file_subjects.append(file_subject)
    
file_subjects


phoenix_uof_metadata = {
    'file_subject': file_subjects,
    'data_filepath': data_filepaths,
    'data_dict_filepath': data_dict_filepaths,
}

# Build a data frame which holds all the City of Phoenix UOF file paths and their corresponding labels
phoenix_uof_metadata_df = pd.DataFrame(phoenix_uof_metadata)
display(phoenix_uof_metadata_df)
```

```python

# to store the mappings, which is intended to map each file_subject to its corresponding data frame
data_dict = {}

for ind in range(len(phoenix_uof_metadata_df)):
    dict_df = pd.read_csv( phoenix_uof_metadata_df['data_dict_filepath'][ind] )
    data_df = pd.read_csv( phoenix_uof_metadata_df['data_filepath'][ind] )
    
    # check and drop '_id' column only for the first df
    if ind == 0 and '_id' in data_df.columns:
        data_df = data_df.drop('_id', axis=1)
    
    # assign column labels based on the descriptive labels, stored in the data dictionary
    data_column_labels = dict_df['label'].tolist()
    data_df.columns = data_column_labels[:len(data_df.columns)]

    # get the file_subject
    file_subject = phoenix_uof_metadata_df['file_subject'][ind]

    # add the data_df to the dictionary with the file_subject as key
    data_dict[file_subject] = data_df
    
    output_path = path_folder + 'relabeled/' + file_subject + '.csv'
    
    # save the relabeled dataframe at the output_path
    data_df.to_csv(output_path, index=False)
```

## Recollect with Updated Data

```python
# obtain all of the file paths from the CPhoenix UOF Data folder and sort them into a data frame by subject

path_folder = 'CPhoenix UOF Data/'
data_filepaths = glob.glob(f'{path_folder}*__.csv', recursive = True)
data_dict_filepaths = glob.glob(f'{path_folder}*-data-dictionary.csv', recursive = True)
data_relabeled_filepaths = glob.glob(f'{path_folder}relabeled/*.csv', recursive = True)

file_subjects = []

for path_data in data_filepaths:
    
    file_subject = path_data.rsplit('\\', 1)[-1][5:-6] 
    file_subjects.append(file_subject)
    
file_subjects


phoenix_uof_metadata = {
    'file_subject': file_subjects,
    'data_filepath': data_filepaths,
    'data_dict_filepath': data_dict_filepaths,
    'data_relabeled_filepath': data_relabeled_filepaths
}

# Build a data frame which holds all the City of Phoenix UOF file paths and their corresponding labels
phoenix_uof_metadata_df = pd.DataFrame(phoenix_uof_metadata)
display(phoenix_uof_metadata_df)
```

```python

# to store the mappings, which is intended to map each file_subject to its corresponding data frame
data_dict = {}

for ind in range(len(phoenix_uof_metadata_df)):
    data_relabeled_df = pd.read_csv( phoenix_uof_metadata_df['data_relabeled_filepath'][ind] )
    
    # get the file_subject
    file_subject = phoenix_uof_metadata_df['file_subject'][ind]

    # add the data_df to the dictionary with the file_subject as key
    data_dict[file_subject] = data_relabeled_df
```

```python
def print_col_info(i):
    print_ln = "+" * 50
    print(f'{print_ln}\n')
    print('CSV NAME:', phoenix_uof_metadata_df['file_subject'][i])
    print()
    print('Columns:\n--------------')
    
    num_columns = len(data_dict[phoenix_uof_metadata_df['file_subject'][i]].columns)
    columnss = data_dict[phoenix_uof_metadata_df['file_subject'][i]].columns
    for col, col_ind in zip(columnss, range(num_columns)):
        print(f"{col_ind + 1}: {col}")
        
    print(f'\n{print_ln}')
        
def preview_csv(i):
    print('\nCSV NAME:', phoenix_uof_metadata_df['file_subject'][i])
    display(data_dict[phoenix_uof_metadata_df['file_subject'][i]])
```

## Clean the Data

```python
file_folder = "CPhoenix UOF Data/relabeled/"
file_name = "use-of-force-incident-details.csv"
file_path = file_folder + file_name

# Read the CSV file
data = pd.read_csv(file_path, na_values=["Not Available", "Not Applicable"])

##### cleaning type 1 -> target rows with NA

# Filter for rows with Year >= 2021
data = data[data['Year'] >= 2021]

# Remove columns that are mostly NA
# - "Individual.Resistance...Other"
# - "Use.of.Force.Utilized...Carotid"
data = data.drop(columns=['Individual Resistance - Other', 'Use of Force Utilized - Carotid'])

# Display the number of NA values for each column
na_count_data = data.isna().sum()
print("------- NA Counts for df 1:\n", na_count_data)

# Remove rows containing NA values
data = data.dropna()

# Replace "Yes" with 1 and "No" with 0
data = data.replace({"Yes": 1, "No": 0})

# Convert columns containing 0s and 1s to integer type
for col in data.columns:
    if data[col].isin([0, 1]).all():
        data[col] = data[col].astype(int)

# View the modified df
# print(data)

# Results in 4.9K rows and 55 cols left
print("Data Dimensions (cleaned):", data.shape)

data.to_csv(file_folder + "cleaned_1-uof-incident-details.csv", index=False)

##### cleaning type 2 -> target columns with NA

data_2 = pd.read_csv(file_path, na_values=["Not Available", "Not Applicable"])

remove_NA_cols = [0, 10, 11, 12, 14, 30, 34] + list(range(37, 45)) + [52, 55]

data_2 = data_2.drop(columns=data_2.columns[remove_NA_cols])

# Display the number of NA values for each column
na_count_data_2 = data_2.isna().sum()
print("\n\n------- NA Counts for df 2:\n", na_count_data_2)
data_2 = data_2.dropna()

# Replace "Yes" with 1 and "No" with 0
data_2 = data_2.replace({"Yes": 1, "No": 0})

# Convert columns containing 0s and 1s to integer type
for col in data_2.columns:
    if data_2[col].isin([0, 1]).all():
        data_2[col] = data_2[col].astype(int)

# View the modified df
# print(data_2)

# Results in 8.6K rows and 40 cols left
print("Data Dimensions (cleaned):", data_2.shape)

data_2.to_csv(file_folder + "cleaned_2-uof-incident-details.csv", index=False)
```
